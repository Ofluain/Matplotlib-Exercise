```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
agelist = [20,30,50,55,59,78,80,82,91]
wightlist = [21,35,53,56,60,66,74,79,82]
```

```python
plt.figure(figsize=(14,8))
plt.plot(agelist,wightlist,"ro-")
plt.xlabel("Yaş")
plt.ylabel("Kilo")
plt.title("Kilonun Yaşa Oranlarının Grafiği")
plt.show()
```

```python
plt.figure(figsize=(12,8)) # boyutları birlikte yazılır yani figsize = 100 kat alınır 
```

```python
x = np.arange(7)
y = x**3 / 4 
y1 = y*2

plt.plot(x,y, "bo-")
plt.plot(x,y1,"ro-")
```

```python
x1= [20,30,50,55,59,78,80,82,91]
y2 = [21,35,53,56,60,66,74,79,82]

plt.figure(figsize = (6,5))
plt.plot(x1,y2, "g")
plt.scatter(x1,y2 , color = "red")

plt.show()
```

```python
x = np.arange(100) / 10
y = np.sin(x) 
y1 = np.cos(x)



plt.scatter(x,y1, color = "red", label = "Cosx" )
plt.plot(x,y, color = "blue", label = "Sinx" )

plt.legend(loc = "best")

plt.show()
```

## Matpotlib-Barplot & Histogram 

```python
data = {
    "Yıl" : np.arange(2010 , 2021),
    "Satış" : np.random.randint(100,500,11),
    "Masraf" : np.random.randint(50,400,11)
}


df = pd.DataFrame(data)
df

df["Bölgeler"]= np.random.choice(["Kuzey", "Güney" , "Doğu" , "Batı"], df.shape[0])

df
```

```python
plt.plot(df["Yıl"] , df["Satış"] , 
         linestyle = "--" , label = "Satışlar",
         alpha = 0.9, color = "blue",marker = "o",
         markerfacecolor = "red"
) 


plt.plot(df["Yıl"] , df["Masraf"] ,
         linestyle = "-", color = "black",
         marker = "x", label = "Masraf"
)



plt.title("Yıllara göre satış ve masraf oranları" )
plt.xlabel("Satışlar")
plt.ylabel("Yıllar")


plt.legend(loc = "best")
plt.show()
```

## plt.bar() (sütun şeklinde grafik olayıdır)

```python
a = np.random.randint(10,60,5)
b = ["Metehan", "Hasan" , "ALi" , "Gamze", "Berfin"] 

plt.bar(b,a , color = "red" ,
        width=0.5,edgecolor = "blue", linewidth = 4,
)
```

```python
data = {
    "Şehir" : ["Ankara" , "İstanbul" , "İzmir" , "Trbzon"],
    "Nüfus" : np.random.randint(30000 , 100000 , 4)

}
```
